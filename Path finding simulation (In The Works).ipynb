```python
def is_inside_the_square(x, y):
    if(-1 < x < 1 and -1 < y < 1):
        return True
    else:
        return False


def is_inside_the_circle(C, R, x, y):
    if((x-C[0])**2+(y-C[1])**2<(R**2)):
        return True
    else:
        return False
    

#circle1
C=(0.3,0.2) # center
R=0.06       # radius
#circle2
C1=(0.33,0.38)
R1=0.03

d = 0.01 #fixed 

angle = 45

poss_angles=[-5,+5]

n = 100
number_of_random_walks = 100
#initial position

fig = plt.figure()
ax = fig.add_subplot(111, aspect='equal')

list_of_most_steps = []

for i in range(number_of_random_walks):
    x = [0]
    y = [0]
    list_of_steps = []
    for i in range(n):
        list_of_steps.append([x1,y1])
        angle = angle + random.choice(poss_angles)
        x1 = x[-1]+d*math.cos(math.radians(angle))
        x.append(x1)
        y1 = y[-1]+d*math.sin(math.radians(angle))
        y.append(y1)
        if(is_inside_the_circle(C1, R1, x1, y1)):
            break
        if(is_inside_the_circle(C, R, x1, y1)):
            break
        if(is_inside_the_square(x1, y1)):
            continue
        else:
            break
    ax.plot(x,y,'',linewidth=0.2)

ax.add_patch(patches.Rectangle((-1, -1),2.0,2.0,fill=False))    
ax.add_patch(patches.Circle(C,R,color='k',fill=True))
ax.add_patch(patches.Circle(C1,R1,color='k',fill=True)) 
plt.axis([-1.5, 1.5, -1.5, 1.5])
plt.show()
```
